```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('//home//ishaan//Documents//Work//Data//Raw Data.csv', error_bad_lines = False, sep = ';')
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 5000)
```

```python
df['Career Drawn'].replace(
    to_replace=['indian army', 'commando', 'army', 'soldier', 'Army officer'],
    value='Army',
    inplace=True
)

#Army = df[(df['Career Drawn'] == 'indian army' ) | (df['Career Drawn'] == 'commando' ) | (df['Career Drawn'] == 'army' ) | (df['Career Drawn'] == 'soldier') | (df['Career Drawn'] == 'Army officer' ) ]
#df.at[Army.index, 'Career Final'] = 'Army'
```

```python
df['Career Drawn'].replace(
    to_replace=['teacher'],
    value='Teacher',
    inplace=True
)

df['Career Drawn'].replace(
    to_replace=['lawyer'],
    value='Lawyer',
    inplace=True
)
df['Career Drawn'].replace(
    to_replace=['engineer'],
    value='Engineer',
    inplace=True
)
```

```python
df['Career Drawn'].replace(
    to_replace=['doctor / nurse', 'doctor'],
    value='Doctor',
    inplace=True
)
```

```python
Teachers = df[(df['Career Drawn'] == 'Teacher' ) | (df['Career Drawn'] == 'Professor' )]
df.at[Teachers.index, 'Career Final'] = 'Teacher'
```

```python
doc = df[(df['Career Drawn'] == 'doctor / nurse') |(df['Career Drawn'] == 'dentist') | (df['Career Drawn'] == 'doctor') | (df['Career Drawn'] == 'Doctor')  ]
df.at[doc.index, 'Career Final'] = 'Doctor'
```

```python
df['Career Final'] = ""
```

```python
doc = df[(df['Career Drawn'] == 'Driver') |  (df['Career Drawn'] == 'ambulance driver') | (df['Career Drawn'] == 'driver') | (df['Career Drawn'] == 'Bus Driver')  ]
df.at[doc.index, 'Career Final'] = 'Drivers and Mobile Plant Operators'
```

```python
df['Gender of the drawing'].replace(
    to_replace=['female'],
    value='Female',
    inplace=True
)
df['Gender of the drawing'].replace(
    to_replace=['male'],
    value='Male',
    inplace=True
)
df['Gender'].replace(
    to_replace=['boy'],
    value='Boy',
    inplace=True
)
df['Gender'].replace(
    to_replace=['girl'],
    value='Girl',
    inplace=True
)
```
