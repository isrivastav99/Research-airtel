```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/home/is99/Documents/Py/Research-airtel/Data/data.csv', error_bad_lines=False, sep = ',')
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 5000)
df.drop(df.columns[27:], axis =1, inplace = True)
```

```python
imp_index = ['Sno', 'Gender' ,'Name of the Student', 'Class', 'Career Drawn' , 'Gender of the drawing']
```

```python
df['Gender of the drawing'] = df['Gender of the drawing'].str.capitalize()
df['Gender'] = df['Gender'].str.capitalize()
df['Career Drawn'] = df['Career Drawn'].str.capitalize()
df['Favourite Colour'] = df['Favourite Colour'].str.capitalize()
```

```python
df['Career Drawn'].replace(
    to_replace=['indian army', 'commando', 'soldier', 'Army officer'],
    value='Army',
    inplace=True
)

#Army = df[(df['Career Drawn'] == 'indian army' ) | (df['Career Drawn'] == 'commando' ) | (df['Career Drawn'] == 'army' ) | (df['Career Drawn'] == 'soldier') | (df['Career Drawn'] == 'Army officer' ) ]
#df.at[Army.index, 'Career Final'] = 'Army'

df['Career Drawn'].replace(
    to_replace=['doctor / nurse', 'doctor'],
    value='Doctor',
    inplace=True
)

Teachers = df[(df['Career Drawn'] == 'Teacher' ) | (df['Career Drawn'] == 'Professor' )]
df.at[Teachers.index, 'Career Final'] = 'Teacher'

doc = df[(df['Career Drawn'] == 'doctor / nurse') |(df['Career Drawn'] == 'dentist') | (df['Career Drawn'] == 'doctor') | (df['Career Drawn'] == 'Doctor')  ]
df.at[doc.index, 'Career Final'] = 'Doctor'
```

```python
df['Gender of the drawing'].replace(
    to_replace=['female', 'femake'],
    value='Female',
    inplace=True
)
df['Gender of the drawing'].replace(
    to_replace=['male'],
    value='Male',
    inplace=True
)
df['Gender'].replace(
    to_replace=['boy'],
    value='Boy',
    inplace=True
)
df['Gender'].replace(
    to_replace=['girl'],
    value='Girl',
    inplace=True
)
```
